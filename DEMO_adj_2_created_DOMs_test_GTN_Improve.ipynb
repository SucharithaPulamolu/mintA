# The MDD model 

```python
import torch
import torch.nn as nn  # Ensure this is included
import torch.nn.functional as F
from torch_geometric.nn import TransformerConv, Linear

class GraphTransformerNetwork(torch.nn.Module):
    def __init__(self, metadata, input_channels, hidden_channels, out_channels, num_layers =3, num_heads =8, dropout=0.05):
        super(GraphTransformerNetwork, self).__init__()

        # Initial linear layer to project input features to hidden dimension
        self.input_linear = Linear(input_channels, hidden_channels)

        # TransformerConv layers
        self.layers = torch.nn.ModuleList()
        for _ in range(num_layers):
            conv = TransformerConv(
                in_channels=hidden_channels,
                out_channels=hidden_channels // num_heads,  # Split across heads
                heads=num_heads,
                dropout=dropout
            )
            self.layers.append(conv)

        # Final linear layer for classification
        self.lin = Linear(hidden_channels, out_channels)

        self.metadata = metadata
        self.hidden_channels = hidden_channels
        self.out_channels = out_channels
        self.num_layers = num_layers
        self.num_heads = num_heads
        self.dropout = nn.Dropout(dropout)

    def forward(self, x_dict, edge_index_dict):
        # Extract specific node and edge data
        x = x_dict['domain_node']
        edge_index = edge_index_dict[('domain_node', 'apex', 'domain_node')]

        # Apply input linear layer
        x = self.input_linear(x)
        x = F.relu(x)  # Apply activation function

        # Apply each TransformerConv layer
        for layer in self.layers:
            x = layer(x, edge_index)  # Apply transformer layer
            x = F.leaky_relu(x)       # Activation function
            x = self.dropout(x)       # Apply dropout for regularization

        # Apply the final linear layer for classification
        out = self.lin(x)
        return out

# Initialize the model
model = GraphTransformerNetwork(
    metadata=data.metadata(),
    input_channels=45,    # Input feature size
    hidden_channels=128,  # Hidden dimension size
    out_channels=2,       # Binary classification
    num_layers=3,         # Number of TransformerConv layers
    num_heads=4,          # Number of attention heads
    dropout=0.1           # Dropout rate
).to(device)
```

```python
# Import necessary libraries
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.utils import dense_to_sparse

# Define the Graph Transformer Network (GTN) surrogate model
class SurrogateGTN(nn.Module):
    def __init__(self, in_features, hidden_features, out_features, num_layers, num_heads, dropout=0.1):
        super(SurrogateGTN, self).__init__()
        self.layers = nn.ModuleList()
        for _ in range(num_layers):
            self.layers.append(
                nn.TransformerEncoderLayer(
                    d_model=hidden_features,
                    nhead=num_heads,
                    dropout=dropout,
                    activation='gelu'
                )
            )
        self.input_proj = nn.Linear(in_features, hidden_features)
        self.output_proj = nn.Linear(hidden_features, out_features)

    def forward(self, features, edge_index=None):
        x = self.input_proj(features)
        for layer in self.layers:
            x = layer(x)
        return self.output_proj(x)

# Set device for CUDA compatibility
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Data to train the surrogate
train_mask = data['domain_node'].train_mask
labels_train = data['domain_node'].y.to(device)  # Move labels to GPU if available
labelled_labels_train_lox = np.where(labels_train.cpu() < 2)
lox_train = np.where(train_mask.cpu() > 0)[0]

labels = data['domain_node'].y.to(device)
lox_train_space = labelled_labels_train_lox
adv_nodes_train = lox_train_space[0][:4000]

# Extract adjacency and features for surrogate model
Adj_sur = extract_As_jan(data, adv_nodes_train)
features_sur = data['domain_node'].x[adv_nodes_train].to(device)  # Ensure features are on the same device

# Convert adjacency matrix to edge_index format for GTN compatibility
edge_index_sur, _ = dense_to_sparse(torch.tensor(Adj_sur, dtype=torch.float32).to(device))

# Define training indices
idx_train = torch.arange(len(adv_nodes_train)).to(device)

# Initialize the GTN surrogate model
surrogate = SurrogateGTN(
    in_features=features_sur.shape[1],  # Input feature size
    hidden_features=128,  # Increased hidden features for better representation
    out_features=2,  # Number of output classes
    num_layers=3,  # Number of Transformer layers
    num_heads=4,  # Number of attention heads
    dropout=0.1  # Dropout for regularization
).to(device)

# Define optimizer and loss function
optimizer_surrogate = torch.optim.Adam(surrogate.parameters(), lr=0.005, weight_decay=0.001)
loss_fn = nn.CrossEntropyLoss()

# Train the GTN surrogate model
surrogate.train()
for epoch in range(50):  # Adjust the number of epochs as needed
    optimizer_surrogate.zero_grad()
    out_surrogate = surrogate(features_sur)  # Forward pass (edge_index_sur not used currently)
    loss = loss_fn(out_surrogate[idx_train], labels[adv_nodes_train][idx_train])  # Compute loss
    loss.backward()
    optimizer_surrogate.step()
    if epoch % 10 == 0:  # Print loss every 10 epochs
        print(f"Epoch {epoch}, Loss: {loss.item()}")

# The GTN surrogate model is now trained and ready for further adversarial testing
```

```python
def one_trial(adv_nodes_test, A_adv):
    # Set device for compatibility
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # Initialize and load dataset
    dataset = DNS2('myGraph_datasets/DNS', transform=T.Compose([T.NormalizeFeatures(), T.ToUndirected()]), balance_gt=True)
    data = dataset[0]

    # Load and assign features
    feats_new = torch.load('feats_new2.pt')
    if feats_new.shape[0] < data.x_dict['domain_node'].shape[0]:
        raise ValueError("The tensor 'feats_new' has fewer rows than the number of domain nodes.")

    data['domain_node'].x = feats_new[:data.x_dict['domain_node'].shape[0], :]
    data['ip_node'].x = torch.zeros(data['ip_node'].x.shape[0], 1)
    data['host_node'].x = torch.zeros(data['host_node'].x.shape[0], 1)
    del feats_new

    # Validate tensors
    for key, value in data.x_dict.items():
        if torch.isnan(value).any() or torch.isinf(value).any():
            raise ValueError(f"The tensor {key} contains NaN or Inf values.")
        if value.shape[0] == 0:
            raise ValueError(f"The tensor {key} has no rows.")

    # Ensure edge indices are valid
    for key, edge_index in data.edge_index_dict.items():
        data.edge_index_dict[key] = edge_index.long()
        valid_mask = (edge_index[0] < data[key[0]].x.shape[0]) & (edge_index[1] < data[key[2]].x.shape[0])
        data.edge_index_dict[key] = edge_index[:, valid_mask]

    # Move tensors to float32
    for key in data.x_dict:
        data[key].x = data[key].x.float()

    # Move data to the device
    try:
        data = data.to(device)
        print(f"Moved data to {device} successfully.")
    except RuntimeError as e:
        print(f"Error moving data to {device}: {e}")
        raise

    print(f"Data device: {data['domain_node'].x.device}")
    x = data['domain_node'].x.cpu()

    perf_arr = np.empty((6, 0))
    val_nodes = np.array([1, 5, 20, 40, 60, 80, 100])

    # Initialize and train a new Graph Transformer Network model
    input_channels = data['domain_node'].x.shape[1]
    model0 = GraphTransformerNetwork(
        metadata=data.metadata(),
        input_channels=input_channels,
        hidden_channels=128,
        out_channels=2,
        num_layers=3,
        num_heads=4,
        dropout=0.1
    ).to(device)

    data.x_dict2 = copy.deepcopy(data.x_dict)
    data.edge_index_dict2 = copy.deepcopy(data.edge_index_dict)

    # Train the model
    for epoch in range(201):
        loss = train(model0, data)

    labels_sur = model_qurey(model0, data, idx_train)
    labels_sur = np.array(labels_sur.cpu())

    # Generate valid indices for adversarial nodes
    max_nodes = data['domain_node'].x.shape[0]
    adv_nodes_test = adv_nodes_test[adv_nodes_test < max_nodes]

    if len(adv_nodes_test) == 0:
        print(f"adv_nodes_test is empty after filtering. Re-generating...")
        adv_nodes_test = np.random.choice(max_nodes, size=min(100, max_nodes), replace=False)

    adv_nodes_test = np.unique(adv_nodes_test)

    if len(adv_nodes_test) == 0:
        raise ValueError("adv_nodes_test has no valid indices within bounds.")

    preds = preds_of_adv(model0, data, adv_nodes_test)

    # Initialize and train a new surrogate model using GTN
    features_sur = data['domain_node'].x[:100]
    Adj_sur = A_adv[:100, :100]
    surrogate = SurrogateGTN(
        in_features=features_sur.shape[1],
        hidden_features=64,
        out_features=2,
        num_layers=2,
        num_heads=4,
        dropout=0.1
    ).to(device)

    features_sur_cpu = features_sur.to(device)
    Adj_sur_cpu = torch.tensor(Adj_sur, dtype=torch.float32).to(device)
    idx_train_cpu = np.random.choice(
        features_sur_cpu.shape[0], size=min(100, features_sur_cpu.shape[0]), replace=False
    )
    labels_sur_cpu = torch.tensor(labels_sur[idx_train_cpu], dtype=torch.long).to(device)

    # Define optimizer and loss function for GTN
    optimizer_surrogate = torch.optim.Adam(surrogate.parameters(), lr=0.005, weight_decay=0.001)
    loss_fn = nn.CrossEntropyLoss()

    # Train GTN surrogate model
    surrogate.train()
    for epoch in range(50):
        optimizer_surrogate.zero_grad()
        out_surrogate = surrogate(features_sur_cpu, Adj_sur_cpu)
        loss = loss_fn(out_surrogate[idx_train_cpu], labels_sur_cpu)
        loss.backward()
        optimizer_surrogate.step()
        if epoch % 10 == 0:
            print(f"Surrogate Epoch {epoch}, Loss: {loss.item()}")

    # Loop over val_nodes and calculate performance metrics
    for val in val_nodes:
        temp0 = data.edge_index_dict2[('domain_node', 'apex', 'domain_node')].cpu().numpy()
        X = x[adv_nodes_test, :].cpu().numpy()
        A = A_adv
        X2 = copy.deepcopy(X)

        simi_arr = np.zeros([100, 100])
        messages = np.dot(A, X2)

        for i in range(100):
            tempx = A[i, :]
            js = np.flatnonzero(tempx)
            for j in range(100):
                simi_arr[i, j] = np.linalg.norm(X[i] + np.mean(messages[js], axis=0))

        k = min(val * val, simi_arr.size)
        row_indices, col_indices = largest_indices(simi_arr, k)

        m = np.zeros([100, 100])
        for i in range(len(row_indices)):
            m[row_indices[i], col_indices[i]] = 1

        A2 = do_perturb_adj(A, m, device)
        aa = A_to_edge_index(torch.tensor(A2, dtype=torch.int64), device)
        conv = np.zeros_like(aa.cpu())
        for k in range(aa.shape[1]):
            conv[0, k] = adv_nodes_test[aa[0, k]]
            conv[1, k] = adv_nodes_test[aa[1, k]]

        all_edges = temp0
        adv_edge_lox = np.nonzero(np.in1d(all_edges[0, :], adv_nodes_test))[0]
        non_adv_edges = np.delete(all_edges, adv_edge_lox, axis=1)
        temp2 = np.hstack((non_adv_edges, conv))

        data.edge_index_dict2[('domain_node', 'apex', 'domain_node')] = torch.tensor(temp2, dtype=torch.int64).to(device)

        ASRgood, num_of_1_forced_to_0, num_of_1, ASRbad, num_of_0_forced_to_1, num_of_0 = calc_ASR(data, adv_nodes_test, model0)

        temp_res = np.array([ASRgood, num_of_1_forced_to_0, num_of_1, ASRbad, num_of_0_forced_to_1, num_of_0]).reshape(-1, 1)
        perf_arr = np.hstack((perf_arr, temp_res))

    return perf_arr
```
